```python
import fastbook
from fastbook import *
from fastai.vision import *
import pandas as pd
import matplotlib as mpl
import numpy as np
```

```python
project_path  = ('C:/Users/conno/Documents/pokemonClassifier/pokemonimages')
```

```python
p_path = Path(project_path)
```

```python
np.random.seed(42)
```

```python
data = ImageDataLoaders.from_folder(p_path, valid_pct=0.2, item_tfms=Resize(460), batch_tfms=aug_transforms(size=224))
data.show_batch()
```

```python
learn = cnn_learner(data, resnet34, metrics=error_rate)
learn.fit_one_cycle(8)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_top_losses(9, figsize=(15,10))
```
